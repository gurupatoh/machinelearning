### merge dataframes with relationships

```python
df=pd.merge(df1,df2, on="question_id" ,how='inner')
```

```python
df
```

```python
dataframe = pd.merge(df3, df, on=['person_id','year'] , how='inner')
```

```python
dataframe
```

```python
dataframe.describe()
```

```python
print(dataframe.groupby(['region_id','year'])[['score']].sum())
```

### obtaining scores band and distributions

```python
final=dataframe.groupby(['region_id','year'])[['score']].describe()
print(final)
```

```python
# percentile list 
perc =[.20, .40, .60, .80] 

global_scores=dataframe.groupby('year')[['score']].describe(percentiles =perc)
print(global_scores)
```

```python
final = pd.read_excel('final.xlsx')
```

### stacked Bar graph showing score distribution by region

```python

# plot data in stack manner of bar type 
final.plot(x='count', kind='bar', stacked=True, 
        title='score distribution by region(2017/2018/2019)')
```
